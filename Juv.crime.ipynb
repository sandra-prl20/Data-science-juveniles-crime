```python
# Calculate total crimes by gender and age group
age_groups = ['Boys 7-12 Years', 'Girls 7-12 Years', 'Boys 12-16 Years', 'Girls 12-16 Years', 'Boys 16-18 Years', 'Girls 16-18 Years']
total_by_age_group = juv_age[age_groups].sum()
```

```python

# Plotting Age Group vs Gender
total_by_age_group.plot(kind='bar', color=['blue', 'pink', 'blue', 'pink', 'blue', 'pink'])
plt.title('Total Crimes by Age Group and Gender')
plt.xlabel('Age Group and Gender')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
# Trends over time by gender
df_year_gender = juv_age.groupby('Year').agg({'Total for boys all Age Groups': 'sum', 'Total for girls all Age Groups': 'sum'})
df_year_gender.plot(kind='line')
plt.title('Trends in Juvenile Crime Rates Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
# State-wise total crimes
statewise_total = juv_age.groupby('Area_Name').agg({'Grand total': 'sum'}).sort_values('Grand total', ascending=False)
statewise_total.plot(kind='bar', figsize=(10,6))
plt.title('Total Juvenile Crimes by State/UT')
plt.xlabel('Area_Name')
plt.ylabel('Number of Crimes')
plt.show()
```

```python

# Gender differences across states
statewise_gender_diff = juv_age.groupby('Area_Name').agg({'Total for boys all Age Groups': 'sum', 'Total for girls all Age Groups': 'sum'})
statewise_gender_diff.plot(kind='bar', stacked=True, figsize=(12,7), color=['blue', 'pink'])
plt.title('Gender Differences in Juvenile Crimes Across States')
plt.xlabel('Area_Name')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
# Step 2: Merge the datasets on 'Area_Name' and 'Year' and 'Sub_Group_Name'
df_merged = pd.merge(juv_fam, juv_edu, on=['Area_Name', 'Year', 'Sub_Group_Name'], how='inner')
df_merged = pd.merge(df_merged, juv_reci, on=['Area_Name', 'Year', 'Sub_Group_Name'], how='inner')
df_merged = pd.merge(df_merged, juv_age, on=['Area_Name', 'Year'], how='inner')
```

```python
print(juv_age['CRIME'].unique())
```

```python
# Example of aggregating data by age group
df_agg = juv_age.groupby(['Area_Name', 'Year']).agg({
    'Boys 7-12 Years': 'sum',
    'Girls 7-12 Years': 'sum',
    'Boys 12-16 Years': 'sum',
    'Girls 12-16 Years': 'sum',
    'Boys 16-18 Years': 'sum',
    'Girls 16-18 Years': 'sum',
    'Total for boys all Age Groups': 'sum',
    'Total for girls all Age Groups': 'sum',
    'Grand total': 'sum'
}).reset_index()

# Plot distribution of crimes by age group
plt.figure(figsize=(12, 6))
df_agg[['Boys 7-12 Years', 'Girls 7-12 Years', 'Boys 12-16 Years', 'Girls 12-16 Years', 'Boys 16-18 Years', 'Girls 16-18 Years']].plot(kind='bar', stacked=True)
plt.title('Crime Distribution by Age Group')
plt.xlabel('Area_Namw and Year')
plt.ylabel('Crime Count')
plt.legend(title='Age Group')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_agg, x='Year', y='Grand total', hue='Area_Name', marker='o')
plt.title('Total Crime Trends Over Time by Area_Name')
plt.xlabel('Year')
plt.ylabel('Total Crime Count')
plt.legend(title='Area_Name')
plt.show()
```
